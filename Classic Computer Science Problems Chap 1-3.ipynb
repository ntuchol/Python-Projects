```python
def Fibonacci(n):
    if n<= 0:
        print("Incorrect input")
    # First Fibonacci number is 0
    elif n == 1:
        return 0
    # Second Fibonacci number is 1
    elif n == 2:
        return 1
    else:
        return Fibonacci(n-1)+Fibonacci(n-2)

# Driver Program

print(Fibonacci(10))
```

```python
class BitSequence:
    def __init__(self, value: int, num_bits: int):
        if not isinstance(value, int):
            raise TypeError("Value must be an integer.")
        if not isinstance(num_bits, int) or num_bits <= 0:
            raise ValueError("Number of bits must be a positive integer.")
        
        self._value = value & ((1 << num_bits) - 1)
        self._num_bits = num_bits

    @property
    def value(self) -> int:
        return self._value

    @property
    def num_bits(self) -> int:
        return self._num_bits
    
    def __len__(self) -> int:
        return self._num_bits

    def __getitem__(self, index: int) -> int:
         if not isinstance(index, int):
            raise TypeError("Index must be an integer.")
         if not -self._num_bits <= index < self._num_bits:
            raise IndexError("Index out of range.")

         if index < 0:
            index += self._num_bits

         return (self._value >> index) & 1

    def __setitem__(self, index: int, bit: int):
        if not isinstance(index, int):
            raise TypeError("Index must be an integer.")
        if not 0 <= index < self._num_bits:
            raise IndexError("Index out of range.")
        if bit not in (0, 1):
             raise ValueError("Bit must be 0 or 1.")

        if bit == 1:
            self._value |= (1 << index)
        else:
            self._value &= ~(1 << index)

    def __int__(self) -> int:
        return self._value

    def __repr__(self) -> str:
        return f"BitSequence(value={self._value}, num_bits={self._num_bits})"

    def __str__(self) -> str:
         return bin(self._value)[2:].zfill(self._num_bits)
```

```python
import time
import random

# Generate a list of 1 million numbers
data = list(range(1000000))
target_value = random.choice(data) # Pick a random number to search for

# Linear Search
def linear_contains(data, target):
    for item in data:
        if item == target:
            return True
    return False

# Binary Search (Requires a sorted list)
def binary_contains(data, target):
    low = 0
    high = len(data) - 1
    while low <= high:
        mid = (low + high) // 2
        if data[mid] == target:
            return True
        elif data[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return False

# Time the linear search
start_time = time.time()
linear_search_result = linear_contains(data, target_value)
end_time = time.time()
linear_time = end_time - start_time

# Time the binary search
start_time = time.time()
binary_search_result = binary_contains(data, target_value)
end_time = time.time()
binary_time = end_time - start_time

# Print results
print(f"Linear Search Result: {linear_search_result}")
print(f"Binary Search Result: {binary_search_result}")
print(f"Linear Search Time: {linear_time:.4f} seconds")
print(f"Binary Search Time: {binary_time:.4f} seconds")
```

```python
import random
from collections import deque
import heapq

def generate_maze(rows, cols, p=0.3):
    maze = [['#' for _ in range(cols)] for _ in range(rows)]
    for i in range(rows):
        for j in range(cols):
            if random.random() > p:
                maze[i][j] = ' '
    maze[1][1] = 'S'
    maze[rows - 2][cols - 2] = 'E'
    return maze

def print_maze(maze):
    for row in maze:
        print(''.join(row))

def get_neighbors(maze, row, col):
    rows, cols = len(maze), len(maze[0])
    neighbors = []
    if row > 0 and maze[row - 1][col] != '#':
        neighbors.append((row - 1, col))
    if row < rows - 1 and maze[row + 1][col] != '#':
        neighbors.append((row + 1, col))
    if col > 0 and maze[row][col - 1] != '#':
        neighbors.append((row, col - 1))
    if col < cols - 1 and maze[row][col + 1] != '#':
        neighbors.append((row, col + 1))
    return neighbors

def dfs(maze, start, end):
    stack = [start]
    visited = {start}
    count = 0
    while stack:
        count += 1
        curr = stack.pop()
        if curr == end:
            return count
        for neighbor in get_neighbors(maze, *curr):
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)
    return count

def bfs(maze, start, end):
    queue = deque([start])
    visited = {start}
    count = 0
    while queue:
        count += 1
        curr = queue.popleft()
        if curr == end:
            return count
        for neighbor in get_neighbors(maze, *curr):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return count

def heuristic(node, end):
    return abs(node[0] - end[0]) + abs(node[1] - end[1])

def astar(maze, start, end):
    open_set = [(heuristic(start, end), start)]
    closed_set = set()
    came_from = {}
    g_score = {start: 0}
    f_score = {start: heuristic(start, end)}
    count = 0

    while open_set:
        count += 1
        curr = heapq.heappop(open_set)[1]
        if curr == end:
            return count

        closed_set.add(curr)

        for neighbor in get_neighbors(maze, *curr):
            if neighbor in closed_set:
                continue
            tentative_g_score = g_score[curr] + 1

            if neighbor not in g_score or tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = curr
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = tentative_g_score + heuristic(neighbor, end)
                heapq.heappush(open_set, (f_score[neighbor], neighbor))
    return count

if __name__ == "__main__":
    num_mazes = 100
    maze_size = 20
    dfs_counts = []
    bfs_counts = []
    astar_counts = []

    for _ in range(num_mazes):
        maze = generate_maze(maze_size, maze_size)
        start = (1, 1)
        end = (maze_size - 2, maze_size - 2)

        dfs_counts.append(dfs(maze, start, end))
        bfs_counts.append(bfs(maze, start, end))
        astar_counts.append(astar(maze, start, end))

    print(f"DFS Counts: {dfs_counts}")
    print(f"BFS Counts: {bfs_counts}")
    print(f"A* Counts: {astar_counts}")

    print(f"Average DFS Count: {sum(dfs_counts) / num_mazes}")
    print(f"Average BFS Count: {sum(bfs_counts) / num_mazes}")
    print(f"Average A* Count: {sum(astar_counts) / num_mazes}")
```

```python
from collections import deque

def is_valid(state):
    missionaries_left, cannibals_left, boat_side = state
    missionaries_right = total_missionaries - missionaries_left
    cannibals_right = total_cannibals - cannibals_left

    if missionaries_left < 0 or cannibals_left < 0 or missionaries_right < 0 or cannibals_right < 0:
        return False
    if (missionaries_left > 0 and missionaries_left < cannibals_left) or \
       (missionaries_right > 0 and missionaries_right < cannibals_right):
        return False
    return True

def get_next_states(state):
    missionaries_left, cannibals_left, boat_side = state
    next_states = []
    moves = [(1, 0), (0, 1), (2, 0), (0, 2), (1, 1)]

    for move in moves:
        delta_missionaries, delta_cannibals = move
        new_boat_side = 1 - boat_side

        if boat_side == 0:
            new_state = (missionaries_left - delta_missionaries, cannibals_left - delta_cannibals, new_boat_side)
        else:
            new_state = (missionaries_left + delta_missionaries, cannibals_left + delta_cannibals, new_boat_side)

        if is_valid(new_state):
            next_states.append(new_state)
    return next_states

def solve_missionaries_and_cannibals(num_missionaries, num_cannibals):
    global total_missionaries, total_cannibals
    total_missionaries = num_missionaries
    total_cannibals = num_cannibals
    initial_state = (num_missionaries, num_cannibals, 0)
    goal_state = (0, 0, 1)
    visited = {initial_state}
    queue = deque([(initial_state, [])])

    while queue:
        current_state, path = queue.popleft()
        if current_state == goal_state:
            return path + [current_state]

        for next_state in get_next_states(current_state):
            if next_state not in visited:
                visited.add(next_state)
                queue.append((next_state, path + [current_state]))
    return None

num_missionaries = 4
num_cannibals = 4

solution = solve_missionaries_and_cannibals(num_missionaries, num_cannibals)

if solution:
    print(f"Solution for {num_missionaries} missionaries and {num_cannibals} cannibals:")
    for state in solution:
        print(state)
else:
    print(f"No solution found for {num_missionaries} missionaries and {num_cannibals} cannibals.")
```

```python
def find_empty_location(board):
    for row in range(9):
        for col in range(9):
            if board[row][col] == 0:
                return (row, col)
    return None

def is_valid(board, num, pos):
    for i in range(9):
        if board[pos[0]][i] == num and pos[1] != i:
            return False
        if board[i][pos[1]] == num and pos[0] != i:
            return False
    box_x = pos[1] // 3
    box_y = pos[0] // 3
    for i in range(box_y * 3, box_y * 3 + 3):
        for j in range(box_x * 3, box_x * 3 + 3):
            if board[i][j] == num and (i,j) != pos:
                return False
    return True

def solve_sudoku(board):
    find = find_empty_location(board)
    if not find:
        return True
    else:
        row, col = find
    for num in range(1,10):
        if is_valid(board, num, (row, col)):
            board[row][col] = num
            if solve_sudoku(board):
                return True
            board[row][col] = 0
    return False

# Example Sudoku puzzle (0 represents empty cells)
board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

if solve_sudoku(board):
    for row in board:
        print(row)
else:
    print("Solution does not exist")
```
